This program is a sideprogram for the image enhancer super resolution GAN. \
It resizes the image dataset into low and high resolution images, which are used for training and testing the GAN.\

The dataset is taken from: http://press.liacs.nl/mirflickr/mirdownload.html

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# This is the directory of the image dataset
train_dir = "image dataset/mirflickr25k"
```

```python
img_ar = cv2.imread("image dataset/mirflickr25k/original_images/im10897.jpg")
print(img_ar)
```

```python

for img in os.listdir("image dataset/mirflickr25k/original_images"):
    print("Image name: ", img)

    img_array = cv2.imread("image dataset/mirflickr25k/original_images/" + img)
    
    # print("Image array: ", img_array)
    img_array = cv2.resize(img_array, (128,128))
    lr_img_array = cv2.resize(img_array,(32,32))
    cv2.imwrite("image dataset/mirflickr25k/hr_images/" + img, img_array)
    cv2.imwrite("image dataset/mirflickr25k/lr_images/"+ img, lr_img_array)
```
